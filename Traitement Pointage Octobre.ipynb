```python
from datetime import datetime
# Fonction permettant de convertir les heures en secondes
def convertionSecondes(data):
    pointage = data["Pointage"]
    depointage = data["Dépointage"]
    
    for i in range(0, len(pointage)):
        heurePointage = pointage[i]
        heureDepointage = depointage[i]
        try:
            #Pointage
            secondesPointage = datetime.strptime(heurePointage, "%H:%M:%S").time()
            secondesPointage = secondesPointage.hour * 3600 + secondesPointage.minute * 60 + secondesPointage.second
            secondesDepointage = datetime.strptime(heureDepointage, "%H:%M:%S").time()
            secondesDepointage = secondesDepointage.hour * 3600 + secondesDepointage.minute * 60 + secondesDepointage.second
            if(secondesDepointage < secondesPointage):
                p1 = pointage[i]
                p2 = depointage[i]
                pointage[i] = p2
                depointage[i] = p1
                print(i)
        except Exception as e:
            print("Erreur de formatage au niveau de la position {}".format(i))
            break
convertionSecondes(data)
```
