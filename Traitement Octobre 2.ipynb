```python
from datetime import datetime
# Fonction permettant de convertir les heures en secondes
def convertionSecondes(data):
    pointage = data["Pointage"]
    depointage = data["Dépointage"]
    
    for i in range(0, len(pointage)):
        heurePointage = pointage[i]
        heureDepointage = depointage[i]
        try:
            #Pointage
            secondesPointage = datetime.strptime(heurePointage, "%H:%M:%S").time()
            secondesPointage = secondesPointage.hour * 3600 + secondesPointage.minute * 60 + secondesPointage.second
            secondesDepointage = datetime.strptime(heureDepointage, "%H:%M:%S").time()
            secondesDepointage = secondesDepointage.hour * 3600 + secondesDepointage.minute * 60 + secondesDepointage.second
            if(secondesDepointage < secondesPointage):
                p1 = pointage[i]
                p2 = depointage[i]
                pointage[i] = p2
                depointage[i] = p1
                print(i)
        except Exception as e:
            print("Erreur de formatage au niveau de la position {}".format(i))
            break
convertionSecondes(data)
```

```python
# Representation des sites suivant des indexes
sites = list(set(data["POINTEUSE"]))
representationSite = []
for i in range(0, len(sites)):
    representationSite.append({"index" : "{}".format(i), "short" : "{}".format(sites[i]), "desc" : "{}".format(sites[i])})
representationSite.append({"index" : "21", "short": "Other", "desc": "Other"})
print(representationSite)
# Ecriture dans un fichier
with open("sites.txt", "w") as f:
    f.write(str(representationSite))
```

```python
# Convertion des heures en minutes
heuresPointage = list(set(data["Pointage"]))
heuresDepointage = list(set(data["Dépointage"]))
for i in heuresPointage:
    timeSplit = i.split(":")
    data["Pointage"][data["Pointage"] == i] = int(timeSplit[0]) * 60 + int(timeSplit[1])
for i in heuresDepointage:
    timeSplit = i.split(":")
    data["Dépointage"][data["Dépointage"] == i] = int(timeSplit[0]) * 60 + int(timeSplit[1])
data
```

```python
# Représentation des pointeuses par leurs indexes
# for j in representationSite:
#     print(list(j.values()))
for i in sites:
    for j in representationSite:
        if(i in list(j.values())):
            data["POINTEUSE"][data["POINTEUSE"] == i] = list(j.values())[0]
data
```

```python
# Creation du fichier .tsv
import csv

with open('octobre.tsv', 'wt') as out_file:
    tsv_writer = csv.writer(out_file, delimiter='\t')
    tsv_writer.writerow(['month'])
    
# Représentation d'un user sur une même ligne 
idUsers = list(set(data["ID"]))
for j in idUsers:
    user = data[data["ID"] == j]
    user.index = range(0, len(user))

    ligne = "{},{},{}".format(user.loc[0]["Pointage"], user.loc[0]["Dépointage"], user.loc[0]["POINTEUSE"])
    for i in range(1, len(user)):
        ligne = ligne + ",{},{},{}".format(user.loc[i]["Pointage"], user.loc[i]["Dépointage"], user.loc[i]["POINTEUSE"])
    with open('octobre.tsv', 'a') as out_file:
        tsv_writer = csv.writer(out_file, delimiter='\t')
        tsv_writer.writerow([ligne])
```
