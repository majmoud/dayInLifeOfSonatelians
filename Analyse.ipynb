# Analyse des données

```python
import pandas as pd
```

```python
f_novembre = pd.read_csv("pointage_novembre.csv", sep=",")
f_octobre = pd.read_csv("pointage_octobre.csv", sep=",")
```

```python
print(f_novembre)
```

```python
print(f_octobre)
```

```python
# Vérification de l'existence de colonnes vide
liste_colonnes = f_octobre.columns
liste_colonnes
```

```python
f_octobre.isnull().sum()
```

```python
f_novembre.isnull().sum()
```

```python
# Calcul du pourcentage de données manquantes pour le mois d'Octobre
pourcentageO = f_octobre.shape[0] / f_octobre.isnull().sum()[3]
pourcentageO
```

```python
# Calcul du pourcentage de données manquantes pour le mois de Novembre
pourcentageN = f_novembre.shape[0] / f_novembre.isnull().sum()[3]
pourcentageN
```

```python
# Suppression des lignes contenant NaN
f_octobre = f_octobre.dropna()
f_novembre = f_novembre.dropna()
```

```python
print(f_novembre.isnull().sum())
```

```python
print(f_octobre.isnull().sum())
```

```python
# Enregistrement des fichiers nettoyés
f_octobre.to_csv("pointageOctobre_nettoye.csv", sep=",")
f_novembre.to_csv("pointageNovembre_nettoye.csv", sep=",")
```

```python
# Fusionnage des 2 fichiers Octobre et Novembre
f1 = pd.read_csv("pointageOctobre_nettoye.csv", sep=",")
f2 = pd.read_csv("pointageNovembre_nettoye.csv", sep=",")
fichierFinal = pd.concat([f1, f2])
```
