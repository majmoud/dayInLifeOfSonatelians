# Le principe consiste à calculer la différence entre deux durée avant de l'écrire au niveau du fichier TSV

```python
with open('octobre4.tsv', 'wt') as out_file:
        tsv_writer = csv.writer(out_file, delimiter='\t')
        tsv_writer.writerow(['month'])
        
for j in list(set(data["ID"])):
    user = data[data["ID"] == j]
    user.index = range(0,len(user))

    timeUser = []
    # Heure d'arriver au service
    timeUser.append(random.randrange(420, 510)) #Entre 7h et 8h30
    # Les heures de déplaçement des agents une fois arrivée au service entre 9h et 16h
    for i in range(1, len(user)):
        timeUser.append(random.randrange(540, 960))
    # Heure de descente entre 16h45 et 18h
    timeUser.append(random.randrange(1005, 1080))
    # Trier le time par ordre croissant
    timeUser.sort()

    # Calcul de la différence du time
    diffTime = [timeUser[0]]
    for i in range(1, len(timeUser)):
        diffTime.append(timeUser[i] - timeUser[i-1])

    ligne = "21" # Position Other
    user["POINTEUSE"][len(user)] = 21 #Permet d'ajouter la position other en fin.
    for i in range(0, len(timeUser)):
        ligne = ligne + "," + str(diffTime[i]) + "," + str(user["POINTEUSE"][i])
    # Gérer les 24h
    ligne = ligne + ",{}".format(str(1440 - sum(diffTime)))

    with open('octobre4.tsv', 'a') as out_file:
            tsv_writer = csv.writer(out_file, delimiter='\t')
            tsv_writer.writerow([ligne])
```
